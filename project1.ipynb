## 1. Load Libraries and Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('2015.csv')
df.head()
```

```python
print("Shape of the dataset:", df.shape)
print("\n Data Info:\n")
df.info()
df.describe(include='all')
```

## 2. Cleaning

#### 1. Missing Values

```python
#Missing Values Analysis
missing=pd.DataFrame(
    {
        'missing_count':df.isnull().sum(),
        'missing_percent':(df.isnull().sum()/len(df))*100
    }
)
missing.sort_values(by='missing_percent',ascending=False)
```

```python
# Checking missing values before handling
print("Missing Values before Cleaning:\n", df.isnull().sum())

# Most columns in 2015 World Happiness dataset do not usually have missing values,
# but if any numeric columns have missing values, fill with mean:
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())

# For object type columns (like Country or Region), fill missing with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Drop any columns with too many missing values (none typically in WHR2015, but here's how)
# Example: df.drop(columns=['Column_Name'], inplace=True)

print("Missing Values after Cleaning:\n", df.isnull().sum())
```

```python
import pandas as pd

missing = pd.DataFrame({
    'missing_count': df.isnull().sum(),
    'missing_percent': (df.isnull().sum() / len(df)) * 100
})
missing_sorted = missing.sort_values(by='missing_percent', ascending=False)
print(missing_sorted)
```

```python
# Handling Missing Values (if any)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
print("Missing Values after Cleaning:\n", df.isnull().sum())
```

#### Insights
- We filled important columns and removed highly missing ones.
- This keeps maximum information while making the dataset clean.

#### 2. Duplicates & Datatypes

```python
# Duplicates & Data Types
print("Duplicate rows:", df.duplicated().sum())
df['Country'] = df['Country'].astype('category')
df['Region'] = df['Region'].astype('category')
df.info()
```

## 3. Univariate,Bivariate & Multivariate Analysis

```python
# 3. Univariate, Bivariate & Multivariate Analysis
# Univariate Analysis - Numerical
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
print(df[num_cols].describe())
sns.histplot(df['Happiness Score'], bins=30, kde=True, color="skyblue")
plt.title("Happiness Score Distribution")
plt.show()
```

```python
# Univariate Analysis - Categorical
cat_cols = df.select_dtypes(include=['category']).columns
for col in cat_cols:
    sns.countplot(x=col, data=df)
    plt.title(f"Count of {col}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Bivariate Analysis
sns.scatterplot(x='Economy (GDP per Capita)', y='Happiness Score', data=df)
plt.title("Relationship between GDP and Happiness Score")
plt.show()
```

```python
# Correlation and Multivariate Analysis
corr_cols = ['Happiness Score', 'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity']
corr = df[corr_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 4. Outlier Detection using IQR on 'Happiness Score'
Q1 = df['Happiness Score'].quantile(0.25)
Q3 = df['Happiness Score'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['Happiness Score'] < lower) | (df['Happiness Score'] > upper)]
print("Number of outliers in Happiness Score:", len(outliers))
```

```python
df['Happiness Score_capped'] = df['Happiness Score'].clip(lower, upper)
sns.boxplot(x=df['Happiness Score_capped'])
plt.title("Happiness Score after Outlier Capping")
plt.show()
```

```python
# 5. Feature Engineering (Optional)
# Example: Create a categorical feature for GDP level
df['GDP_Category'] = pd.cut(df['Economy (GDP per Capita)'], bins=[-np.inf, 0.5, 1, np.inf], labels=['Low', 'Medium', 'High'])
sns.boxplot(x='GDP_Category', y='Happiness Score_capped', data=df)
plt.title("Happiness Score by GDP Category")
plt.show()
```

# 6. Key Insights
#### - Economy (GDP per Capita), Family support, Health, and Freedom have strong positive correlations with Happiness Score.
#### - Generosity and Trust in government show weaker correlations.
#### - The capped Happiness Score removes extreme values to reduce skewness impact on analysis.
